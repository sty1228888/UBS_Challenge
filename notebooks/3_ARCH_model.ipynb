**Introduction**

ARCH, introduced by <a href="https://en.wikipedia.org/wiki/Robert_F._Engle">Engle</a> (<a href="https://www.jstor.org/stable/1912773">1982</a>), is an Autoregressive model with Conditional Heteroskedasticity. 

ARCH and GARCH models are highly relevant to ```economic forecasting``` and ```measuring volatility```. ```ARCH``` model is concerned about modeling volatility of the variance. Concerning the mean, both ```stationary and nonstationary``` cases are applicable. 

Some other models adopted are ```ARCH```, ```ARCH-M```, ```GARCH```, ```GARCH-M```, ```TGARCH```, and ```EGARCH```. Our research in this project will focus on the deployment of ```ARCH``` and ```GARCH```. 

**Rationale**

- ```Auto Regressive```: The current value can be expressed as a function of the previous values. 

- ```Conditional```: The variance is based on past errors. 

- ```Heteroskedasticity```: The series displays unusual variance (varying variance).

In real financial sector, the variance may change consistently over time, either increasing or decreasing. 

Compared to other autoregressive models like ```AR```, ```ARIMA``` and ```SARIMA```, ```ARCH``` can model time-varying variance. 

**Mechanism**

- Calculate the ```daily volatility``` by finding the standard deviation of $f(S_i) = ln(\frac{S_i}{S_{i-1}})$, where $S$ is a given variable. 

- Auto-regressive models applies when a correlation can be observed in the variance change over time. 

- The time-lasting factor is a hyper-parameter to be passed to the ARCH model. It will decide how many previous days' various change (errors) can be used to predict today's. For example: 

  - ```ARCH(1)```
    
    $\epsilon(t) = w(t) * \sigma(t) = w(t)* \sqrt{(\alpha_{0} + \alpha_{1} * \epsilon^{2}(t - 1))}$

  - ```ARCH(2)```
    
    $\epsilon(t) = w(t) * \sigma(t) = w(t)* \sqrt{(\alpha_{0} + \alpha_{1} * \epsilon^{2}(t - 1) + \alpha_{2} * \epsilon^{2}(t - 2))}$

  - ```ARCH(n)```
    
    $\epsilon(t) = w(t) * \sigma(t) = w(t)* \sqrt{(\alpha_{0} + \sum_{i=1}^{t} \alpha_{i} * \epsilon^{2}(t - i))}$

**Eligibility**

- The ```ARCH``` model effectiveness is derived from ```squaring the residuals``` and ```examining the correlogram```. 
  
  The ```mean of the residuals should be zero```. 

- Series with significant trends or seasonal effects are not ```ARCH```-applicable. Instead, ```ARIMA``` will be a better choice.
